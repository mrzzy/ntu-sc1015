# Export csv files

This export is used to export the main categories for each business for the model training

```python
import pandas as pd
import seaborn as sb
sb.set_theme()
import re
import os
```

```python
business = pd.read_json(path_or_buf="/content/drive/MyDrive/Colab Notebooks/yelp_academic_dataset_business.json", lines=True)
categories = business["categories"]
```

```python
category_list = ["Active Life", "Arts & Entertainment", "Automotive", "Beauty & Spas", "Education", "Event Planning & Services", "Financial Services", "Food", "Health & Medical", "Home Services", "Hotels & Travel", "Local Flavor", "Local Services",  "Mass Media", "Nightlife", "Pets", "Professional Services", "Public Services & Government", "Religious Organizations", "Restaurants", "Shopping", "Others"]
```

```python
update_category = []
for index, x in enumerate(categories):
  status = False
  if not x:
    update_category.append(category_list[-1])
    status = True

  for category_index, category in enumerate(category_list):
    if status:
      break

    if category in x:
      update_category.append(category_list[category_index])
      status = True
  if not status:
    update_category.append(category_list[-1])
```

```python
export_df = business[["business_id", "state", "categories"]].copy()
export_df.insert(3, "main_category", update_category, True)
export_df.info()
```

```python
path = ("/content/drive/MyDrive/Colab Notebooks/Export/simplified_cat.csv")
export_df.to_csv(path_or_buf=path)
```

```python
states_pd = business["state"]

state_set = set()
for state in states_pd:
  state_set.add(state)
```

```python
category_index = pd.DataFrame()
state_index = pd.DataFrame()
```

```python
category_index.insert(0, "main_category", category_list, True)
state_index.insert(0, "state", list(state_set), True)
```

```python
path = ("/content/drive/MyDrive/Colab Notebooks/Export/category_index.csv")
category_index.to_csv(path_or_buf=path)
```

```python
path = ("/content/drive/MyDrive/Colab Notebooks/Export/state_index.csv")
state_index.to_csv(path_or_buf=path)
```
